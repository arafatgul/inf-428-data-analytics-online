```python
titanic=pd.read_csv('titanic_data.csv')
print(titanic.shape)
print(titanic.head())
```

```python
titanic=titanic.fillna(0)
titanic.head()
```

```python
def convertSex(sex):
    if sex=='female':
        return 0
    else:
        return 1
```

```python
titanic['gender']=titanic.sex.apply(lambda x: convertSex(x))
```

```python
titanic.drop('sex',axis=1)
```

```python
def survivedToString(survived):
    if survived==1:
        return 'survived'
    else:
        return 'perished'
```

```python
titanic['survived']=titanic.survived.apply(lambda x: survivedToString(x))
```
