```python
import pandas as pd
```

```python
# open the draft data
draft=pd.read_csv('nba_draft_combine_all_years.csv', index_col=0)
```

```python
# open the nba data
nba_2020=pd.read_csv('nba_2020_advanced.csv', index_col=0)
```

```python
# take a look at the first few columns of nba 2020 advanced stats data
nba_2020.head()
```

```python
# take a look at the draft data
draft.head()
```

```python
# OK here is the tricky part... to join the data sets we need to make the Player names equivalent
# Notice that the nba 2020 data has name followed by ID... so we use a trick from assignment 3 to reformat
# the name so that it is equivalent to the draft data (ie we get rid of the ID)
nba_2020.Player=nba_2020.Player.astype(str).apply(lambda x:x.split('\\')[0])
draft.Player=draft.Player.astype(str)
```

```python
# now we can join the two data sets 
# see https://pandas.pydata.org/pandas-docs/stable/reference/api/pandas.DataFrame.merge.html
joined=pd.merge(nba_2020, draft, left_on='Player', right_on='Player')
```

```python
# take a look at the size of the joined data, and the first few rows of the joined data
# note there are only 240 players in the joined set... this is because some nba players were drafted before 2009 or 
# made it to the league undrafted, while not all drafted players are in the NBA in 2020
print(nba_2020.shape)
print(draft.shape)
print(joined.shape)
joined.head()
```
